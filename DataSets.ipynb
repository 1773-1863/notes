```python
from itertools import product 

keys = ['sl','sw','pl','pw']

i = 0
N = len(keys)

plt.figure(figsize=(15,15))

for v,w in product(keys, keys):
    i+=1
    plt.subplot(N,N,i)
    if v == w:
        plt.hist(df_iris[v],bins=20)
        plt.xlabel(v)
    else:
        plt.scatter(df_iris[v],df_iris[w],c=df_iris['c'],cmap='prism')
        plt.xlabel(v)
        plt.ylabel(w)
        
plt.show()
```

```python
import re

MAXLINES = 10000
f = open('/Users/cemgil/Dropbox/Public/swe546/data/hamlet.txt')

i = 0
words = []

for l in f:
    i+=1
    #    print(l,end='')
    if i>MAXLINES: break
    for w in re.split('(\W+)',l):
        if w.isalpha():
            words.append(w)
    
        
f.close()

Counts = dict()

for w in words:
    if Counts.has_key(w):
        Counts[w] += 1
    else:
        Counts[w] = 1
        
for w in sorted(Counts.keys()):
    if w.isupper() and len(w)>1:
        print(w, Counts[w])

for w in words:
    if w.isupper() and len(w)>1:
        print(w)
```

## Wind Tribune Data

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pylab as plt
import pandas as pd

plt.figure(figsize=(14,5))
df_windtribune = pd.read_csv(u'data/wind_tribune.csv')
plt.plot(df_windtribune.wind_speed)
plt.show()
```

Since one of the dimensions corresponds to an angle, it is natural to visualize the data using a radar plot.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
r = df_windtribune['wind_speed']
theta = 2 * np.pi * df_windtribune['wind_direction']/360.
area = 100 * np.abs(df_windtribune['production'])/df_windtribune['production'].max()
#colors = theta

ax = plt.subplot(111, projection='polar')
c = ax.scatter(theta, r, alpha=0.9, s=area, color='b', edgecolor='w')
c.set_alpha(0.75)

plt.show()
```
